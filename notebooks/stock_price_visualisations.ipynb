### Retrieving data from yfinance API

* The first function takes in ticker object, start_date end_date and returns the historical dataframe from the API
* Second function takes in a data_frame with price data and creates a plot of closing price and volume

```python
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

def get_price_history(ticker, start_date, end_date):
    price_history = ticker.history(start=start_date, end=end_date)
    return price_history


def plot_closing_price_volume(data_frame):
    plt.figure(figsize=(16, 12))
    plt.figure(figsize=(16, 12))
    closing_plot = data_frame.Close.plot(legend=True)
    volume_plot = data_frame.Volume.plot(kind='area', secondary_y=True,
                                          alpha=0.6, legend=True)

    closing_plot.margins(0, 0)

    closing_plot.yaxis.set_tick_params(labelsize=18)
    closing_plot.xaxis.set_tick_params(labelsize=18)
    volume_plot.yaxis.set_tick_params(labelsize=18)
    closing_plot.set_ylabel("Closing price", fontsize=20)
    closing_plot.set_xlabel("Date", fontsize=20)
    volume_plot.set_ylabel("Volume (Shares traded)", fontsize=20)
    volume_plot.margins(0, 0)
    plt.grid(True)
    plt.title("Price and volume during analysed period", fontsize=20)
    plt.show()
```

```python

sns.set_style("darkgrid")
gme_ticker = "GME"
gme = yf.Ticker(gme_ticker)
gme_price_history = get_price_history(gme, "2020-12-01", "2021-05-01")
plot_closing_price_volume(gme_price_history)
```

Chose the GameStop price and volume during the end of 2020 - first half of 2021 period values to illustrate this function

Large Spike in Volume, fllowed by the spike in price. Interesting to observe the second spike in traded
volume, and how, the price, although around $\frac{1}{2}$ of its peak, is still very much above where it started
at the end of May

* The parameters above that are passed to get_price_history can be changed - similar to the ticker object and we can observe other stocks.
